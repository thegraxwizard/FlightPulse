```python
# ===================================================================
# === SCRIPT: TRANSFORMER AUTOENCODER FOR ANOMALY DETECTION =======
# ===================================================================

# === Imports ===
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, dayofweek, concat_ws, count, avg, first
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler, StandardScaler
from pyspark.ml.clustering import GaussianMixture # Or KMeans if you prefer
import pandas as pd

# === Step 1: Ensure SparkSession and Base Data ===
try:
    spark
    df_transformed
except NameError:
    print("SparkSession or df_transformed not found. Initializing...")
    spark = SparkSession.builder.appName("FlightPulse-TransformerAE").getOrCreate()
    df = spark.read.csv("flights.csv", header=True, inferSchema=True)
    df_transformed = df.withColumn(
        "flight_date_formatted", to_date(col("FLIGHT_DATE"), "dd-MM-yyyy")
    ).withColumn(
        "day_of_week", dayofweek(col("flight_date_formatted"))
    ).withColumn(
        "route", concat_ws("-", col("DEPARTURE_STATION_CD"), col("ARRIVAL_STATION_CD"))
    ).withColumn( # Add TotalPax for feature consistency
        "TotalPax", col("TIER1_ELIGIBLE_PAX") + col("TIER2_ELIGIBLE_PAX") + col("TIER3_ELIGIBLE_PAX")
    )
    print("--- SparkSession and Base DataFrame Initialized ---")

# === Step 2: Re-run Clustering to get Cluster IDs ===
# (Ensure route_dna and clustering pipeline/model are available or recalculated)
route_dna = df_transformed.groupBy("route").agg(
    count("*").alias("FlightFrequency"), first("HAUL").alias("HaulType"),
    avg(col("FIRST_CLASS_SEATS") + col("BUSINESS_CLASS_SEATS") + col("ECONOMY_SEATS")).alias("AvgTotalSeats"),
    (avg(col("FIRST_CLASS_SEATS") + col("BUSINESS_CLASS_SEATS")) / (avg(col("FIRST_CLASS_SEATS") + col("BUSINESS_CLASS_SEATS") + col("ECONOMY_SEATS")) + 1e-6)).alias("PremiumSeatRatio"),
    (avg(col("TIER1_ELIGIBLE_PAX")) / (avg(col("TIER1_ELIGIBLE_PAX") + col("TIER2_ELIGIBLE_PAX") + col("TIER3_ELIGIBLE_PAX")) + 1e-6)).alias("Tier1PassengerRatio")
)
categorical_cols = ["HaulType"]
numerical_cols = ["FlightFrequency", "AvgTotalSeats", "PremiumSeatRatio", "Tier1PassengerRatio"]
indexer = StringIndexer(inputCol="HaulType", outputCol="HaulType_index")
assembler = VectorAssembler(inputCols=["HaulType_index"] + numerical_cols, outputCol="features_unscaled")
scaler = StandardScaler(inputCol="features_unscaled", outputCol="features")
pipeline_clustering = Pipeline(stages=[indexer, assembler, scaler])
route_ml_ready = pipeline_clustering.fit(route_dna).transform(route_dna)
optimal_k = 9 # Based on GMM BIC
gmm_final = GaussianMixture(featuresCol='features', k=optimal_k, seed=1)
final_model = gmm_final.fit(route_ml_ready)
clustered_routes = final_model.transform(route_ml_ready).select("route", "prediction")
df_with_clusters = df_transformed.join(clustered_routes, on="route")
print("--- Clustering applied ---")

# === Step 3: Prepare Sequences for Target Cluster (Autoencoder) ===
target_cluster_id = 0 # Choose a cluster to train on
X_cluster_data = []
# NOTE: Autoencoders don't need a separate 'y' label; they learn to reconstruct 'X'
seq_length = 10

sequence_pd_base = df_with_clusters.select(
    "route", "flight_date_formatted", "day_of_week", "TotalPax",
    col("prediction").alias("cluster_id")
).orderBy("route", "flight_date_formatted").toPandas()
print("--- Base data converted to Pandas ---")

print(f"--- Filtering data for Cluster {target_cluster_id} ---")
grouped = sequence_pd_base.groupby('route')
for name, group in grouped:
    if group['cluster_id'].iloc[0] == target_cluster_id:
        features = list(zip(group['day_of_week'], group['TotalPax'])) # Features: day, pax
        if len(features) >= seq_length: # Need at least seq_length points
            # Create overlapping sequences
            for i in range(len(features) - seq_length + 1):
                X_cluster_data.append(features[i : i + seq_length])

if not X_cluster_data:
    print(f"Error: No data found for Cluster {target_cluster_id}. Cannot train model.")
else:
    X_cluster_data = np.array(X_cluster_data)

    # Normalize features
    max_pax_cluster = X_cluster_data[:, :, 1].max() if X_cluster_data.size > 0 else 1
    X_cluster_normalized = X_cluster_data.astype(float)
    if max_pax_cluster > 0: X_cluster_normalized[:, :, 1] /= max_pax_cluster

    print(f"--- Data prepared for Cluster {target_cluster_id} Autoencoder (X shape: {X_cluster_normalized.shape}) ---")

    # Split data (use a portion for training, keep rest for anomaly testing later)
    X_train_ae, X_test_ae = train_test_split(X_cluster_normalized, test_size=0.2, random_state=42)

    # === Step 4: Define Transformer Autoencoder Architecture ===
    # Encoder part uses Transformer blocks
    # Decoder part tries to reconstruct the original sequence
    input_shape_ae = X_train_ae.shape[1:] # (seq_length, num_features) = (10, 2)

    def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
        x = layers.MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(inputs, inputs)
        x = layers.Dropout(dropout)(x)
        x = layers.LayerNormalization(epsilon=1e-6)(inputs + x); res = x
        x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(res)
        x = layers.Dropout(dropout)(x)
        x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
        encoded = layers.LayerNormalization(epsilon=1e-6)(res + x)
        return encoded

    # Simple Decoder using TimeDistributed Dense layers
    def build_decoder(encoded_input, original_seq_length, original_num_features):
        # Repeat the encoded vector for each time step to feed into TimeDistributed
        x = layers.RepeatVector(original_seq_length)(encoded_input)
        # Use TimeDistributed Dense layers to reconstruct the sequence step-by-step
        x = layers.TimeDistributed(layers.Dense(64, activation='relu'))(x)
        x = layers.TimeDistributed(layers.Dense(original_num_features))(x) # Output shape matches input
        return x
        
    # Build the Autoencoder model
    inputs = keras.Input(shape=input_shape_ae)
    # Encoder: Apply Transformer block(s)
    encoded = transformer_encoder(inputs, head_size=32, num_heads=2, ff_dim=4, dropout=0.1)
    # We might need a layer to flatten or pool before the decoder
    # For simplicity, let's try GlobalAveragePooling first
    encoded_pooled = layers.GlobalAveragePooling1D(data_format="channels_last")(encoded)

    # Decoder: Reconstruct the sequence
    decoded = build_decoder(encoded_pooled, input_shape_ae[0], input_shape_ae[1])

    autoencoder = keras.Model(inputs, decoded)
    autoencoder.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3), loss='mse')
    autoencoder.summary()

    # === Step 5: Train the Autoencoder ===
    print(f"\n--- Training Transformer Autoencoder for Cluster {target_cluster_id} ---")
    history_ae = autoencoder.fit(
        X_train_ae, X_train_ae, # Autoencoder learns to reconstruct its input
        epochs=50, # Train longer for reconstruction tasks
        batch_size=32,
        validation_split=0.2,
        verbose=1,
        callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)]
    )
    print("--- Autoencoder Training Complete ---")

    # === Step 6: Evaluate Reconstruction Error ===
    print(f"\n--- Evaluating Autoencoder Reconstruction Error for Cluster {target_cluster_id} ---")
    test_loss = autoencoder.evaluate(X_test_ae, X_test_ae, verbose=0)
    print(f"Test Reconstruction Loss (MSE): {test_loss:.6f}")

    # Plot training history
    plt.figure(figsize=(10, 6))
    plt.plot(history_ae.history['loss'], label='Training Loss')
    plt.plot(history_ae.history['val_loss'], label='Validation Loss')
    plt.title(f'Cluster {target_cluster_id} Autoencoder Training History (Loss)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss (MSE)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # === Step 7: Anomaly Detection (Example) ===
   # === Step 7: Anomaly Detection (Example) (CORRECTED) ===

# --- Imports needed for this cell ---
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # <-- ADD THIS IMPORT

# Calculate reconstruction error for each sample in the test set
reconstructions = autoencoder.predict(X_test_ae)
mse = np.mean(np.power(X_test_ae - reconstructions, 2), axis=(1, 2)) # Error per sequence

# Set a threshold (e.g., based on training loss distribution)
threshold = np.percentile(mse, 95) # Flag the top 5% highest errors as anomalies
print(f"\n--- Anomaly Detection Example ---")
print(f"Reconstruction Error Threshold (95th percentile): {threshold:.6f}")

# Find anomalies
anomalies = mse > threshold
print(f"Found {np.sum(anomalies)} anomalies out of {len(X_test_ae)} test samples.")

# Visualize reconstruction error distribution
plt.figure(figsize=(10, 6))
sns.histplot(mse, bins=50, kde=True) # Now sns is defined
plt.axvline(threshold, color='r', linestyle='--')
plt.title('Distribution of Reconstruction Errors (MSE)')
plt.xlabel('Mean Squared Error')
plt.ylabel('Frequency')
plt.show()
```
